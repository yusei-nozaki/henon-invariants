```python
import numpy as np
import matplotlib.pyplot as plt

# Define axis grid
x = np.linspace(-0.5, 0.5, 100)
y = np.linspace(-0.5, 0.5, 100)
X, Y = np.meshgrid(x, y)
# Z = np.sin(np.sqrt(X**2 + Y**2))

def dx_(x,y,dy,e): # dx^2=(E-V)-dy^2
    return np.sqrt(2*e - x**2 - y**2 - 2*x**2*y + 2/3*y**3 - dy**2)

def J(y,dy): # Assumes x=0 and E = 1/12
    y = np.array(y)
    dy = np.array(dy)
    dx=dx_(0,y,dy,E)

    L0 = y**2/6 + y**3/3 + y**4/2 - (10*y**5)/3 + 2*y**6
    L2 = - (1/12+y**2/2-(5/3)*y**3)*(dx**2-dy**2)
    L4 = (1/4)*(dx**4-6*dx**2*dy**2+dy**4)

    return L4 + L2 + L0/2

E = 1/12        # Set energy
Z = J(X,Y)      # Define levels

# plt.figure(dpi=300)

# Ploting contour
plt.contour(X, Y, Z, 20)
plt.xlabel('$y$')
plt.ylabel('$p_y$')
# plt.xlim(-0.5,0.5)
# plt.ylim(-0.5,0.5)
# plt.text(-0.6,0.4,f'$E=1/12$')
plt.axis('equal')
plt.tight_layout()

# Uncomment to save a high quality png
# plt.savefig('contour-deg4.png', dpi=300)

plt.show()
```
