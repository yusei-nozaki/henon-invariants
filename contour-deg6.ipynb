```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Define axis grid
x = np.linspace(-0.5, 0.5, 100)
y = np.linspace(-0.5, 0.5, 100)
X, Y = np.meshgrid(x, y)
# Z = np.sin(np.sqrt(X**2 + Y**2))

def dx_(x,y,dy,e): # dx^2=(E-V)-dy^2
    # s = random.choice([-1,1])
    return np.sqrt(2*e - x**2 - y**2 - 2*x**2*y + 2/3*y**3 - dy**2)

def J(y,dy,e): # Assumes x=0 and E = 1/12
    y = np.array(y)
    dy = np.array(dy)
    dx=dx_(0,y,dy,e)

    l0 = -0.0540404*y**2 + 0.244108*y**3 - 0.0287879*y**4 - 1.44687*y**5 + 5.01279*y**6 - 8.09697*y**7 + 8.54444*y**8 - 3.59506*y**9

    l2 = 0.0606061*y + 0.162121*y**2 - 0.742424*y**3 + 2.10606*y**4 - 2.75152*y**5 + 2.46667*y**6

    l4 = 0.363636*y - 0.5*y**2 + y**3

    return -l0/2 + l2*(dx**2-dy**2) + l4*(dx**4-6*dx**2*dy**2+dy**4) + (dx**6-15*dx**4*dy**2+15*dx**2*dy**4-dy**6)/6



E = 1/12        # Set energy
Z = J(X,Y,E)    # Define levels

# # Ploting contour
plt.contour(X, Y, Z, 20)
plt.xlabel('$y$')
plt.ylabel('$p_y$')
# plt.xlim(-0.5,0.5)
# plt.ylim(-0.5,0.5)
# plt.title('deg = 6, E = 1/12')
# plt.text(-0.45,0.4,f'$E={E}$')
plt.axis('equal')
plt.tight_layout()

plt.savefig('deg6-contour.png', dpi=300)



plt.show()
```
