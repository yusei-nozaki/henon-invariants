```python
import numpy as np
import matplotlib.pyplot as plt

# Define axis grid
x = np.linspace(-0.5, 0.5, 100)
y = np.linspace(-0.5, 0.5, 100)
X, Y = np.meshgrid(x, y)

def dx(x,y,dy,e): # dx^2=(E-V)-dy^2
    return np.sqrt(2*e - x**2 - y**2 - 2*x**2*y + 2/3*y**3 - dy**2)

def J(y,dy,e): # Assumes x=0 and E = 1/12
    y = np.array(y)
    dy = np.array(dy)
    return (dx(0,y,dy,e)**3-3*dx(0,y,dy,e)*dy**2)/3 + dx(0,y,dy,e)*(y - y**2 + 2*y**3)/2

E = 1/12        # Set energy
Z = J(X,Y,E)    # Define levels

# Ploting contour
plt.contour(X, Y, Z, 20)
plt.xlabel('$y$')
plt.ylabel('$p_y$')
# plt.xlim(-0.5,0.5)
# plt.ylim(-0.5,0.5)
# plt.text(-0.6,0.4,f'$E=1/12$')
plt.axis('equal')
plt.tight_layout()

# Uncomment to save a high quality png
# plt.savefig('contour-deg3.png', dpi=300)

plt.show()
```
